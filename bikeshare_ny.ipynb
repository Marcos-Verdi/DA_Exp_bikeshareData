## EDA

All of the data is from 2017 but we can start analyzing trends by the month, day, and even hour.

```python
df['Start Month'].unique()
```

```python
by_month = df.groupby('Start Month').count()
plt.figure(figsize=(20,5))
sns.barplot(data=by_month, x=by_month.index, y=by_month['Day'])
plt.title('Rides per Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.grid()
```

We have data for only the first 6 months of the year. We can see that from January to June the number of rides increase gradually and more than doubles during this period. This gradual increase can be explained mostly by weather. New York's winter are harsh and snowy which makes it not only hard but even dangerous to move around the city in bikes. So it is only presumable that rides will increase once spring and summer start. It would be useful to see the rest of the year's data and examine thebehavior during the rest of the summer and once autumn and winter approach in oder to better determine the impact of weather on ridership.

```python
by_day = df.groupby('Day of Week').count()
plt.figure(figsize=(20,5))
sns.barplot(data=by_day, x=by_day.index, y=by_day['Day'], order=by_day.sort_values('Day', ascending=False).index)
plt.title('Rides per Day')
plt.xlabel('Day')
plt.ylabel('Count')
plt.grid()
```

Our data shows that during the weekend days ridership declines significantly. This is a strong indicator that bikes are being used as more than just a recreational mean of transportation. NY residents see bikes as an option to susbstitute public transportation on their commutes. We can also observe a pronounced peak on Wednesdays where ridership is about 20% higher than Mondays and Fridays. This may be due to the fact that users have more difficulty engaging in significant physical activities after the weekend (Monday) and may tend to give up on it right before the weekend (Fridays). It is probably much easier to keep the routine in the middle of the week. 

```python
by_hour = df.groupby('Start Hour').count()
plt.figure(figsize=(20,5))
sns.barplot(data=by_hour, x=by_hour.index, y=by_hour['Day'], order=by_hour.sort_values('Day', ascending=False).index)
plt.title('Rides per Hour of the Day')
plt.xlabel('Hour of Day')
plt.ylabel('Count')
plt.grid()
```

```python
plt.figure(figsize=(20,10))
trips_filtered = df.loc[(df['Trip Duration mins'] < 120) & (df['Trip Duration'] > 30)]
sns.boxplot(trips_filtered['Trip Duration mins'])
plt.title('Trip DUration Distribution')
plt.xlabel('Trip Duration')
```

```python
plt.figure(figsize=(20,10))
trips_filtered = df.loc[(df['Trip Duration mins'] < 120) & (df['Trip Duration'] > 30)]
sns.distplot(trips_filtered['Trip Duration mins'])
plt.title('Trip DUration Distribution')
plt.xlabel('Trip Duration')
plt.xticks(range(0,120,5));
```

Our graphs show that most rides last between 5 to 20 minutes, with a mean of about 7 minutes. It must be noted that I have not counted trips that lasted more than 2 hours, which represent about 0.1% and have been treated as outliers, Also, there are not data points for rides that last less than 1 minute. It would be interesting to do some further analysis to determine if a ride of more than the average discourages users from using bikes as a transportation option. 

```python
by_route = df.groupby('Route').count().sort_values('Day', ascending=False).head(50)
plt.figure(figsize=(20,10))
sns.barplot(data=by_route, x=by_route['Day'], y=by_route.index, orient='h')
plt.title('Most Traveled Routes by Ride Count')
plt.xlabel("Number of Rides")
plt.grid()
```

We can observe that the most popular route is ridden almost twice as much as its closest competitor. This suggests that we can do further analysis to determine what factors can cause such a difference and if they can be replicated in order to increase ridership in other routes. We can analyze some of the routes individually in order to discern some patterns.

```python
by_st_start = df.groupby('Start Station').count().sort_values('Day', ascending=False).head(50)
plt.figure(figsize=(20,10))
sns.barplot(data=by_st_start, x=by_st_start['Day'], y=by_st_start.index, orient='h')
plt.title('50 Most Used Origin Stations by Ride Count')
plt.xlabel('Number of Rides')
plt.ylabel('Origin Station')
plt.grid()
```

```python
by_st_end = df.groupby('End Station').count().sort_values('Day', ascending=False).head(50)
plt.figure(figsize=(20,10))
sns.barplot(data=by_st_end, x=by_st_end['Day'], y=by_st_end.index, orient='h')
plt.title('50 Most Used Destination Stations by Ride Count')
plt.xlabel('Number of Rides')
plt.ylabel('Destination Station')
plt.grid()
```

Now that we know which are the most used stations, either as an origin or destination, we can do some further exploration in order to determine other patterns based on date and time information.

```python
plt.figure(figsize=(20,10))
sns.distplot(df['Birth Year'],kde=False)
plt.grid()
```

We can observe that the average user was born in the late 80's and early 90's but that there is a considerable quantity of users born in the 70's and early 80's as well.

## Analyzing the most transitted stations and routes

```python
pershing_sq = df.loc[(df['Start Station'] == 'Pershing Square North') | (df['End Station'] == 'Pershing Square North')]
pershing_sq
```

```python
plt.figure(figsize=(20,10))
sns.distplot(pershing_sq['Start Hour'], bins=25)
plt.title('Distribution of Traffic in Pershing Sq as an Origin Station')
plt.xticks(range(0,26));
```

As expected, we observe a bimodal distribution corresponding to peak hours of the day. However, there seems to be significantly more demand in the evening peak hours than in the morning ones. This is probably due to the fact that Pershing Sq is located in a highly commercial area and therefore is, in its majority, used by people getting out from work than the opposite. 

```python
plt.figure(figsize=(20,10))
sns.histplot(pershing_sq['End Hour'], kde=True, bins=25)
plt.title('Distribution of Traffic in Pershing Sq as a Destination Station')
plt.xticks(range(0,26));
```

The distribution is virtually the same for the the use of Pershing Square as a destination station. We can observe a bimodal distribution where the first peak happens in the morning (people getting to work) and the second around 5 and 6 pm which is possibly caused by commuters using Grand Central Terminal which is right in front of Pershing Square.

## Analysis of Most Transitted Routes

```python
most_transitted_routes = by_route.index[0:9]

plt.figure(figsize=(20,10))

for index,i in enumerate(most_transitted_routes):
    temp = df.loc[df['Route'] == i]
    plt.subplot(5,2,index+1)
    plt.tight_layout()
    sns.histplot(temp['End Hour'], bins=24, kde=True)
    plt.title('Route: '+i)
    plt.xticks(range(0,26))
```

We can observe that the top 10 most transitted routes peak mostly at the same time (around evening peak hours) probably caused by commuter usage but a couple of them have somewhat different patterns. For instance, the Greenwhich Ave & St Charles Street Route, which actually has the same station as an origin and destination, has a pretty even distribution throughout the day. Given the fact that this route has the same station as origin and destination we may infere that these bikes are perhaps being used by a delivery or courier service and hence it has that even distribution throughout the day. The same happens with the W 21st & 9th Ave Route in terms of its distribution but in this case this may be explained by the fact that this is not a commuter heavy route.
